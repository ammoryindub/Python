##### Challenge 1: Guess the Number
a program that generates a random number between 1 and 100 and asks the user to guess the number. Provide feedback to the user (e.g., "Too high" or "Too low") until they guess the correct number. Keep track of the number of guesses and display it when the user guesses correctly.

```python
# import library 

import random

def guess_the_number():
    number = random.randint(1, 100)
    attempts = 0

    while True: # initiate while loop
        guess = int(input("Guess a number between 1 and 100: "))
        attempts += 1

        if guess == number:
            print("Congratulations! You guessed the number correctly.")
            print("Number of attempts:", attempts)
            break
        elif guess < number:
            print("Too low. Try again.")
        else:
            print("Too high. Try again.")

guess_the_number()
```

##### Challenge 2: Fibonacci Sequence
a program that prints the Fibonacci sequence up to a given number of terms. 

**Note: The Fibonacci sequence is a series of numbers in which each number is the sum of the two preceding ones. The sequence typically starts with 0 and 1.**

```python
def fibonacci_sequence(n):
    
    # Check if the number of terms is valid
    if n <= 0:
        print("Please enter a positive integer.")
        return
    
    # Initialize the first two terms of the sequence
    sequence = [0, 1]
    
    # Generate the Fibonacci sequence
    while len(sequence) < n:
        next_number = sequence[-1] + sequence[-2]
        sequence.append(next_number)
    
    # Print the sequence
    for number in sequence:
        print(number, end=" ")

# Get the number of terms from the user
n = int(input("Enter the number of terms in the Fibonacci sequence: "))

# Print the Fibonacci sequence
fibonacci_sequence(n)
```

##### Challenge 3: Palindrome Checker
a program that asks the user for a word or phrase and checks if it is a palindrome. 

**Note: A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, disregarding spaces, punctuation, and capitalization.**

```python
def is_palindrome(word):
    # Remove spaces and convert to lowercase
    word = word.replace(" ", "").lower()
    
    # Check if the word is a palindrome
    if word == word[::-1]:
        return True
    else:
        return False

# Get the word or phrase from the user
word = input("Enter a word or phrase: ")

# Check if it is a palindrome
if is_palindrome(word):
    print("It is a palindrome.")
else:
    print("It is not a palindrome.")
```

##### Challenge 4: Rock, Paper, Scissors Game
a program that allows the user to play a game of Rock, Paper, Scissors against the computer. 

Prompt the user to enter their choice (rock, paper, or scissors), generate a random choice for the computer, and determine the winner based on the game rules.

```python
import random

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "You win!"
    else:
        return "Computer wins!"

def play_game():
    choices = ["rock", "paper", "scissors"]
    
    # Get user's choice
    user_choice = input("Enter your choice (rock, paper, or scissors): ").lower()
    
    # Validate user's choice
    if user_choice not in choices:
        print("Invalid choice. Please try again.")
        return
    
    # Generate computer's choice
    computer_choice = random.choice(choices)
    
    # Print user's and computer's choices
    print("Your choice:", user_choice)
    print("Computer's choice:", computer_choice)
    
    # Determine the winner
    print(determine_winner(user_choice, computer_choice))

# Play the game
play_game()
```

##### Challenge 5: Word Count
a program that reads a file and counts the number of occurrences of each word in the file. 

Display the word count for each word in alphabetical order.

```python
def count_words(filename):
    word_count = {}

    try:
        with open(filename, 'r') as file:
            for line in file:
                words = line.strip().split()
                for word in words:
                    # Remove punctuation and convert to lowercase
                    word = word.strip('.,?!-:;\'"').lower()
                    if word:
                        word_count[word] = word_count.get(word, 0) + 1
    except FileNotFoundError:
        print("File not found.")
        return

    # Sort the word count dictionary by keys
    sorted_word_count = sorted(word_count.items())

    # Display the word count for each word
    for word, count in sorted_word_count:
        print(f"{word}: {count}")

# Example usage: count words in a file named 'sample.txt'
count_words('02-Python-Skills.ipynb')
```
